## Web Automation & Web Scraping: HTML - Tags

### HTML Tags & Elements
- h1 : Tag name
- class = "title" : Attribute
- Homepage : Affected content  
- /h1 : End Tag

#### Web Scraping: Xpath - Syntax
- `//tagName`
- `//tagName[1]`
- `//tagName[@AttributeName="Value"]`       This is the standard syntax

#### Xpath functions
- `contains()`
- `starts-with()`

#### Updated Xpath Syntax
- `//tagName[contains(@AttributeName, "Value")]`

#### Xpath Operators
- `and `
- `or`
- Xpath Syntax : `//tagName[(expression 1) and (expression 2)]`

#### Xpath - Testing Xpath
- Go to: scrapinghub.github.io/xpath-playground/

#### Xpath - Special character
- `/`: Select the children from the node set on the left side of this character

- `//`: Specifies that the matching node set should be located at any level within the document.

- **Note**: if you want the text use `/text()`

- `.`: Specifies the current context should be used (refers to present node)

- `..`: Refers to a parent node

- `*`: A wildcard character that selects all elements or attributes regardless of names

- `./*`: Select all the children nodes considering the current context

- `@`: Select an attribute

- `()`: Grouping an XPath expression

- `[n]`: Inddicates that a node with index 'n' should be selected
